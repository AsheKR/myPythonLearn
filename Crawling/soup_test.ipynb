```python
from bs4 import BeautifulSoup

with open('weekday.html') as f:
    html = f.read()
```

```python
soup = BeautifulSoup(html, 'lxml')
```

```python
for a in soup.select('a.title'):
    print(type(a))
```

```python
for a in soup.select('a.title'):
    print(a)
```

```python
for a in soup.select('a.title'):
    print(a.get_text()) # same as a.text
```

```python
img_list = soup.select('div.thumb img')
```

```python
for img in img_list:
    print(img['src'])
```
