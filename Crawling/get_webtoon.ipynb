```python
import requests as rq

URL_WEEKDAY = 'https://comic.naver.com/webtoon/weekday.nhn'

response = rq.get(URL_WEEKDAY)
```

```python
import re
```

```python
webtoons = list(set(re.findall(r'<a.*?href="(?P<url>.*?)".*class="title".*>(?P<name>.*)</a>', response.text)))
```

### re.DOTALL contain \r\n 

```python
web_url_title = list(map(lambda x: {'title':x[0], 'url_thumbnail':x[1]}, webtoons))
```

```python
web_url_title
```

## [{name: name} {thumbnail: thumbnail}] ex 1)

```python
name = list(set(re.findall(r'<a.*?.*class="title".*>(?P<name>.*)</a>', response.text)))
```

```python
exp = re.compile(r'<div class="thumb".*?<img.*?src="(.*?)".*?</div>', re.DOTALL)
```

```python
thumbnail = list(set(re.findall(exp, response.text)))
```

```python
web_list = list()

for i in range(len(name)):
    dict1 = dict()
    dict1['name'] = name[i]
    dict1['thumbnail'] = thumbnail[i]
    web_list.append(dict1)
```

## [{name:name} {thumbnail: thumbnail}] ex 2)

```python
exp = re.compile(r'<div class="thumb".*?<img.*?src="(.*?)".*?.*?</div>.*?<a.*?title="(.*?)".*?</a>', re.DOTALL)
```

```python
find_url_title = set(list(re.findall(exp, response.text)))
```

```python
web_url_title = list(map(lambda x: {'thumbnail':x[0], 'name':x[1]}, find_url_title))
```

```python
web_url_title
```
